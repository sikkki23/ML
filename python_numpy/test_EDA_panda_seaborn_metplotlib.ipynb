```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv('iris_balanced_150.csv')
print(iris.head())
```

```python
print(iris.columns)
print(iris.shape)
```

```python
print(iris['Species'].value_counts()) # Count of each species
```

```python
iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm', title='Sepal Length vs Width')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='Species', height=4 ) \
    .map(plt.scatter, "SepalLengthCm" , "SepalWidthCm" ) \
        .add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='Species', height=4 ) \
    .map(plt.scatter ) \
        .add_legend()
plt.show()
```

```python
#iris_setosa = iris.loc[iris[iris['Species'] == 'Iris-setosa'] ]
iris_setosa = iris[iris['Species'] == 'Iris-setosa']
iris_virginica = iris[iris['Species'] == 'Iris-virginica']
iris_versicolor = iris[iris['Species'] == 'Iris-versicolor']
plt.plot(iris_setosa['PetalLengthCm'], np.zeros_like(iris_setosa['PetalLengthCm']), 'o', label='Setosa')
plt.plot(iris_virginica['PetalLengthCm'], np.zeros_like(iris_virginica['PetalLengthCm']), 'o', label='Virginica')
plt.plot(iris_versicolor['PetalLengthCm'], np.zeros_like(iris_versicolor['PetalLengthCm']), 'o', label='Versicolor')
plt.legend()
```

```python
sns.FacetGrid(iris, hue='Species', height=5 ) \
    .map(sns.displot,"PetalLengthCm") \
        .add_legend()
plt.show()
```

### Cumulative Distributions Function(CDF)

```python
counts , bin_edges = np.histogram(iris_setosa['PetalLengthCm'], bins=10, density=True)
pdf = counts / sum(counts)  # Probability Density Function
cdf = np.cumsum(pdf)         # Cumulative Distribution Function
plt.plot(bin_edges[1:], pdf, label='PDF')
plt.plot(bin_edges[1:], cdf, label='CDF')


counts2 , bin_edges2 = np.histogram(iris_versicolor['PetalLengthCm'], bins=10, density=True)
pdf2 = counts2 / sum(counts2)  # Probability Density Function
cdf2 = np.cumsum(pdf2)         # Cumulative Distribution Function
plt.plot(bin_edges2[1:], pdf2, label='PDF2')
plt.plot(bin_edges2[1:], cdf2, label='CDF2')

counts3 , bin_edges3 = np.histogram(iris_virginica['PetalLengthCm'], bins=10, density=True)
pdf3 = counts3 / sum(counts3)  # Probability Density Function
cdf3 = np.cumsum(pdf3)         # Cumulative Distribution Function
plt.plot(bin_edges3[1:], pdf3, label='PDF3')
plt.plot(bin_edges3[1:], cdf3, label='CDF3')

plt.legend()
plt.show()
```
